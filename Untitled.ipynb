Задание 3

 Для этого задания необходимо разбить все покупки на энное количество групп “Sale_group” (Допустим маленькие продажи, средние и высокие) на основе Sales
Сгруппировать данные на основе региона и группы продаж (Region, Sale_group). Определить основные тенденции и паттерны. Выделить наиболее «прибыльную» группу.


```python
df['Sale_group'] = pd.qcut(df['Sales'], q=3, labels=['Low', 'Medium', 'High'])
```

```python
df
```

```python
grouped_df = df.groupby(['Region', 'Sale_group'])['Sales'].sum().reset_index()
grouped_df = grouped_df.sort_values(by=['Region', 'Sales'], ascending=[True, False])
```

```python
grouped_df
```

Можно заметить, что регионы - Владивосток, Калиниград, Москва и Урал - приносят больше всего дохода в каждой группе продаж.

Так же в этих регионах группа High приносит больше всего продаж.

Владивосток является наиболее прибыльным регионом.

```python
most_profitable_group = grouped_df.groupby('Sale_group')['Sales'].sum().idxmax()
total_sales_per_group = grouped_df.groupby('Sale_group')['Sales'].sum()

print(f"Наиболее прибыльная группа на общенациональном уровне: {most_profitable_group}")
print("\nСуммарные продажи по группам продаж:")
print(total_sales_per_group)
```

Наиболее прибыльной категорией продаж стала High.
